# Calculating $\eta$: From Code to Physics

You asked how `eta` is being calculated in the code and why we use the specific transformation `2 * (epsH - etaH)` in the notebooks.

This notebook breaks down the exact Python implementation in `inf_dyn_background.py` and derives the mathematical relation to the USR parameter.

## 1. The Code Implementation (`inf_dyn_background.py`)

In our background simulation module, we calculate two quantities labeled `epsH` and `etaH`.

````python
import inspect
from inf_dyn_background import get_derived_quantities

# Let's print the actual source code function to be transparent
print(inspect.getsource(get_derived_quantities))
````

### Dissecting the lines:

The code computes:
```python
etaH = -(-3*z*y - v0*model.dfdx(x)/S**2)/(y*z)
```
Recall the equations of motion in the code:
- `dydT` (Acceleration) $= -3zy - \frac{v_0 V'}{S^2}$
- `y` (Velocity) $\propto \dot{\phi}$
- `z` (Hubble) $\propto H$

So the code's `etaH` calculates:
$$ \text{etaH}_{code} = -\frac{\text{Acceleration}}{\text{Velocity} \times \text{Hubble}} = -\frac{\ddot{\phi}}{H \dot{\phi}} $$

In physics literature, this quantity is often denoted as the **Hubble friction parameter**, $\delta_H$.

$$ \delta_H \equiv -\frac{\ddot{\phi}}{H \dot{\phi}} $$

## 2. Converting to $\eta_{USR}$

The parameter that signals USR (and asymptotes to -6) is the rate of change of $\epsilon$.
$$ \eta_{USR} \equiv \frac{d \ln \epsilon_H}{dN} $$

We calculate this in the notebooks using the identity:
$$ \eta_{USR} = 2(\epsilon_H - \delta_H) $$
calculated as:
```python
eta_USR = 2 * (data['epsH'] - data['etaH'])
```

### Derivation of the Identity
Start with the definition of $\epsilon_H$:
$$ \epsilon_H = \frac{\dot{\phi}^2}{2 M_{pl}^2 H^2} $$

Take the log:
$$ \ln \epsilon_H = 2 \ln \dot{\phi} - 2 \ln H - \ln(2 M_{pl}^2) $$

Differentiate with respect to e-folds $N$ (using $\frac{d}{dN} = \frac{1}{H} \frac{d}{dt}$):
$$ \frac{d \ln \epsilon_H}{dN} = 2 \frac{1}{H} \frac{d \ln \dot{\phi}}{dt} - 2 \frac{1}{H} \frac{d \ln H}{dt} $$

1. The second term is simply $2 \epsilon_H$ (since $\epsilon_H = -\frac{\dot{H}}{H^2} = -\frac{d \ln H}{dN}$).
2. The first term:
$$ \frac{1}{H} \frac{d \ln \dot{\phi}}{dt} = \frac{1}{H} \frac{\ddot{\phi}}{\dot{\phi}} = - \left( -\frac{\ddot{\phi}}{H \dot{\phi}} \right) = -\delta_H $$

Combine them:
$$ \frac{d \ln \epsilon_H}{dN} = 2(-\delta_H) - 2(-\epsilon_H) = 2(\epsilon_H - \delta_H) $$

## 3. Why -6?

In the braking phase ($V \gg K.E.$):
- $\epsilon_H$ is small (potential dominated).
- The field stops due to friction: $\ddot{\phi} \approx -3H\dot{\phi}$.
- This means $\delta_H = -\frac{-3H\dot{\phi}}{H\dot{\phi}} = 3$.

Plugging into our formula:
$$ \eta_{USR} \approx 2(0 - 3) = -6 $$

This confirms that the -6 comes directly from the friction term $\delta_H$ being 3, multiplied by the factor of 2 from the square in $\epsilon \propto \dot{\phi}^2$.
